```python
t = var('t')
m = var('m')
b = var('b')
g = var('g')
x0 = var('x0')
v0x = var('v0x')
y0 = var('y0')
v0y = var('v0y')
x = function('x')(t)
y = function('y')(t)
```

```python
odex = m * diff(x,t,2) == -b*diff(x,t,1)
show(odex)
```

```python
odey = m * diff(y,t,2) == -b*diff(y,t,1)-m*g
show(odey)
```

```python
solx = desolve_laplace(odex, x, ivar=t, ics=[0, x0, v0x])
show(solx)
```

```python
soly = desolve_laplace(odey, y, ivar=t, ics=[0, y0, v0y])
show(soly)
```
