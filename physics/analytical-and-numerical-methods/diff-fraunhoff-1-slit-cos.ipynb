```python
ds = var('ds')
y = var('y')
theta = var('theta')

ds = y*sin(theta)
show(ds)
```

```python
dfi = var('dfi')
lmb = var('lmb')

dfi = (ds/lmb)*2*pi
show(dfi)
```

```python
Ey = var('Ey')
E0 = var('E0')
Ey = E0 * cos(dfi)
show(Ey)
```

```python
Esum = var('Esum')
d = var('d')
Esum = integrate(Ey, (y, - d/2, d/2))
show(Esum)
```

```python
S = var('S')
S = Esum^2
show(S)
```

```python
p1 = S.subs(lmb==1,d==5,E0==1)
plot(p1, (theta, -pi/8, pi/8))
```

```python
p2 = (S.subs(lmb==1, d==10, E0==1))^(1/4)
density_plot(p2, (theta, -pi/4, pi/4), (x, -1, 1), plot_points=100)
```
