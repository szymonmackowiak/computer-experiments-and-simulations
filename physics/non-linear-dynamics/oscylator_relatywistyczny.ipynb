```python
def acc_ho(position):
    k = 1
    m = 1
    acc = -(k/m)*position
    return acc
```

```python
x_rk4 = []
x_rk4.append(A)

vx_rk4 = []
vx_rk4.append(0)

for i in range(1, len(t)):
    kx1 = vx_rk4[i-1]
    kv1 = acc_ho(x_rk4[i-1])
    
    kx2=vx_rk4[i-1]+0.5*dt*kv1
    kv2=acc_ho((x_rk4[i-1]+0.5*dt*kx1))
    
    kx3=vx_rk4[i-1]+0.5*dt*kv2;
    kv3=acc_ho((x_rk4[i-1]+0.5*dt*kx2))
    
    kx4=vx_rk4[i-1]+dt*kv3
    kv4=acc_ho((x_rk4[i-1]+dt*kx3))
    
    x_rk4.append(x_rk4[i-1]+(dt/6)*(kx1+2*kx2+2*kx3+kx4))
    vx_rk4.append(vx_rk4[i-1]+(dt/6)*(kv1+2*kv2+2*kv3+kv4))
    
EK_rk4 = []
EP_rk4 = []
EC_rk4 = []
EI_rk4 = []

for i in range(len(t)):
    EK_rk4.append(0.5*m*vx_rk4[i]**2)
    EP_rk4.append(0.5*k*(x_rk4[i]**2))
    EC_rk4.append(EK_rk4[i]+EP_rk4[i])
    EI_rk4.append(EC_rk4[0])

dH = np.sqrt(np.mean(np.power(np.subtract(EI_rk4, EC_rk4), 2)))    
    
plt.plot(x_rk4)
plt.ylabel('x(t) (RK4)')
plt.xlabel('t')
plt.show()    
    
plt.plot(EC_rk4)
plt.ylabel('EC (RK4)')
plt.xlabel('t')
plt.show()
```

```python
def acc_dho(position, velocity):
    k = 1
    m = 1
    b=0.1
    acc = -(k/m)*position - b*velocity
    return acc
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt

dt=0.00001
A=5
k=1
m=1
f=1;
t=np.arange(0, 100, dt)

def acc_dho(position, velocity):
    k = 1
    m = 1
    b=0.1
    acc = -(k/m)*position - b*velocity
    return acc

x_rk4 = []
x_rk4.append(A)

vx_rk4 = []
vx_rk4.append(0)

for i in range(1, len(t)):
    kx1 = vx_rk4[i-1]
    kv1 = acc_dho(x_rk4[i-1], vx_rk4[i-1])
    
    kx2=vx_rk4[i-1]+0.5*dt*kv1
    kv2=acc_dho((x_rk4[i-1]+0.5*dt*kx1), (vx_rk4[i-1]+0.5*dt*kv1))
    
    kx3=vx_rk4[i-1]+0.5*dt*kv2;
    kv3=acc_dho((x_rk4[i-1]+0.5*dt*kx2), (vx_rk4[i-1]+0.5*dt*kv2))
    
    kx4=vx_rk4[i-1]+dt*kv3
    kv4=acc_dho((x_rk4[i-1]+dt*kx3), (vx_rk4[i-1]+dt*kv3))
    
    x_rk4.append(x_rk4[i-1]+(dt/6)*(kx1+2*kx2+2*kx3+kx4))
    vx_rk4.append(vx_rk4[i-1]+(dt/6)*(kv1+2*kv2+2*kv3+kv4))
    
EK_rk4 = []
EP_rk4 = []
EC_rk4 = []
EI_rk4 = []

for i in range(len(t)):
    EK_rk4.append(0.5*m*vx_rk4[i]**2)
    EP_rk4.append(0.5*k*(x_rk4[i]**2))
    EC_rk4.append(EK_rk4[i]+EP_rk4[i])
    EI_rk4.append(EC_rk4[0])

dH = np.sqrt(np.mean(np.power(np.subtract(EI_rk4, EC_rk4), 2)))    
    
plt.plot(x_rk4)
plt.ylabel('x(t) (RK4)')
plt.xlabel('t')
plt.show()    
    
plt.plot(EC_rk4)
plt.ylabel('EC (RK4)')
plt.xlabel('t')
plt.show()
```

```python
def acc_rho(position, velocity, k, m, c):
    om_pow_2 = k/m
    epsilon = (3/2)*(om_pow_2/c)**2
    
    acc = -om_pow_2*position + epsilon*position*velocity**2
    return acc
```

```python
dt=0.001
A=5
k=1
m=1
f=1;
t=np.arange(0, 100, dt)

k = 1
m = 1
c = 1

x_rk4 = []
x_rk4.append(A)

vx_rk4 = []
vx_rk4.append(0)

for i in range(1, len(t)):
    kx1 = vx_rk4[i-1]
    kv1 = acc_rho(x_rk4[i-1], vx_rk4[i-1], k, m, c)
    
    kx2=vx_rk4[i-1]+0.5*dt*kv1
    kv2=acc_rho((x_rk4[i-1]+0.5*dt*kx1), (vx_rk4[i-1]+0.5*dt*kv1), k, m, c)
    
    kx3=vx_rk4[i-1]+0.5*dt*kv2;
    kv3=acc_rho((x_rk4[i-1]+0.5*dt*kx2), (vx_rk4[i-1]+0.5*dt*kv2), k, m, c)
    
    kx4=vx_rk4[i-1]+dt*kv3
    kv4=acc_rho((x_rk4[i-1]+dt*kx3), (vx_rk4[i-1]+dt*kv3), k, m, c)
    
    x_rk4.append(x_rk4[i-1]+(dt/6)*(kx1+2*kx2+2*kx3+kx4))
    vx_rk4.append(vx_rk4[i-1]+(dt/6)*(kv1+2*kv2+2*kv3+kv4))
    
EK_rk4 = []
EP_rk4 = []
EC_rk4 = []
EI_rk4 = []

for i in range(len(t)):
    m_rel = m/(np.sqrt(1-(vx_rk4[i]/c)**2))
    p_rel = m_rel*vx_rk4[i]
    EK_rk4.append(0.5*m_rel*vx_rk4[i]**2)
    EP_rk4.append(0.5*k*(x_rk4[i]**2))
    #EC_rk4.append(EK_rk4[i]+EP_rk4[i])
    EC_rk4.append(np.sqrt((p_rel*c)**2 + (m*c**2)**2) + 0.5*k*(x_rk4[i]**2))
    EI_rk4.append(EC_rk4[0])

dH = np.sqrt(np.mean(np.power(np.subtract(EI_rk4, EC_rk4), 2)))    
    
#x_100p = x_rk4.copy()
#v_100p = vx_rk4.copy()
#x_50p = x_rk4.copy()
#v_50p = vx_rk4.copy()
#x_10p = x_rk4.copy()
#v_10p = vx_rk4.copy()
x_1p = x_rk4.copy()
v_1p = vx_rk4.copy()
    
    
plt.plot(x_rk4)
plt.title('$x(t)$')
plt.ylabel('$x(t)$')
plt.xlabel('$t$')
plt.show()   

plt.plot(vx_rk4)
plt.title('$v_x(t)$')
plt.ylabel('$v_x(t)$')
plt.xlabel('$t$')
plt.show() 
    
plt.plot(x_rk4, vx_rk4)
plt.title('PS')
plt.ylabel('$px(t)$')
plt.xlabel('$x(t)$')
plt.show()    
    
plt.plot(EC_rk4)
plt.ylabel('EC (RK4)')
plt.xlabel('t')
plt.show()
```
