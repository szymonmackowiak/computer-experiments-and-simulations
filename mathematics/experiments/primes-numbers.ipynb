```python
import numpy as np
import matplotlib.pyplot as plt

#ta funcja sprawdza czy number jest liczba pierwsza
def is_prime(number):
    prime = 0
    if number > 1:
       # check for factors
        for i in range(2,int(number**0.5 +1)):
            if (number % i) == 0:
               prime = 0
               break
            else:
               prime = 1
       

    else:
       prime = 0

    return prime

#ta funcja rysuje okregi
def create_circle(r):
    circle = plt.Circle((0, 0), radius = r, fill=False, lw=0.1)
    return circle




#n oznacza liczbe podzialow okregu
for n in range(2, 100):

    c=[] #tablica okregow
    x=[] #tablica wspolrzednych x
    y=[] #tablica wspolrzednych y

    x0=[] #srodek ukladu
    y0=[] #srodek ukladu
    
    alfa = 360/n #kat podzialu okregu
    nmax = 100 #liczba pierscieni
    rmax = nmax #promien najdalszego pierscienia

    for i in range(0,rmax):
        c.append(create_circle(1+i))
        ax = plt.gca()
        ax.add_patch(c[i-1])
        plt.axis('scaled')
        

    for i in range(0,n):
        x.append(rmax*np.cos((np.pi/180)*alfa*i))
        y.append(rmax*np.sin((np.pi/180)*alfa*i))
        x0.append(0)
        y0.append(0)
        plt.plot([x0[i], x[i]], [y0[i], y[i]], '-r')


    #numer pierwscienia dla danej liczby int(19/6)+1

    for i in range(0, int(1*n*nmax)):
        if is_prime(i) == 1:
            r =(int(i/n)+0.5)
            xp=(r*np.cos((np.pi/180)*(alfa)*i+(np.pi/180)*(0.5*alfa)))
            yp=(r*np.sin((np.pi/180)*(alfa)*i+(np.pi/180)*(0.5*alfa)))
            plt.plot(xp, yp, 'ok')

    plt.title("n = "+ str(n) +" podzialow")
    plt.axis('off')
    plt.savefig("n"+str(n)+".jpg") #zapis obrazka
    plt.clf()
```

```python
np1tab = []
np2tab = []

neg_p1tab = []
neg_p2tab = []

neg_np1tab = []
neg_np2tab = []

p1tab = []
p2tab = []

for n in range(1, 10000000):
    p1 = 6*n+1
    p2 = 6*n-1
    
    if (is_prime(p1)):
        np1tab.append(n)
        p1tab.append(p1)
    else:
        neg_np1tab.append(n)
        neg_p1tab.append(p1)
        
        
    if (is_prime(p2)):
        np2tab.append(n)
        p2tab.append(p2)
    else:
        neg_np2tab.append(n)
        neg_p2tab.append(p2)
    #print(f"p1 = {p1}, p2 = {p2}")
    
    if(n%100000 == 0):
        print(f"{n/10000000 * 100}%")
        
```
