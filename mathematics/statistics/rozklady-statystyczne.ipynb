# ROZKŁADY STATYSTYCZNE

## 1. Zmienna losowa

<b>Przypomnienie oznaczeń:</b>

- $E$ - przestrzeń zdarzeń elementarnych 
- $e_i$ - zdarzenie elementarne
- $A_i$, $B_i$ - zdarzenia
- $P$ - prawdopodobieństwo
- $S$ - ciało (pole) zdarzeń - rodzina podzbiorów przestrzeni zdarzeń elementarnych, w którym możliwe jest tworzenie sum, iloczynów, różnic, zdarzeń przeciwnych, pewnych i niemożliwych, dla wszystkich zdarzeń należących do tego zbioru
- $(E, S, P)$ - przestrzeń probabilistyczna związana z danym doświadczeniem losowym

<b>Zmienna losowa</b> - zmienna, która w wyniku doświadczenia może przyjąć wartość z pewnego zbioru liczb rzeczywistych z określonym prawdopodobieństwem. <b>Zmienną losową</b> nazywamy każdą funkcję mierzalną określoną na przestrzeni zdarzeń elementarnych $E$ i przybierającą wartość ze zbioru liczb rzeczywistych $(X:E \rightarrow R)$. Inaczej, jeżeli $(E, S, P)$ jest przestrzenią probabilistyczną dla danego doświadczenia, to zmienną losową jest każda funkcja o wartościach rzeczywistych, określona na zbiorze zdarzeń elementarnych $E$, mierzalną względem ciała zdarzeń $S$, na którym określona jest miara $P$. 

<b>Przykłady: </b>

- liczba elementów wyprodukowana na danym stanowisku w czasie jednej zmiany, 
- ilość energii elektrycznej zużytej przez dane gospodarstwo domowe danego dnia
- wartości cech jednostek statystycznych wybieranych z populacji generalnej
- wyniki pomiarów

Zmienne losowe oznaczamy dużymi literami alfabetu łacińskiego, np. $X, Y, Z,$ ..., natomiast ich realizacje (konkretne wartości) odpowiadającymi im małymi literami $x, y, z,$ ... Wzajemne przyporządkowanie zmiennych losowych i zdarzeń jest <b>jednoznaczne</b>.

<b>Typy zmiennej losowej:</b>

- zmienne typu <b>skokowego (dyskretnego)</b> - zbiór możliwych realizacji jest skończony (liczba ziaren w kłosie, liczba prosiąt w miocie)
- zmienne typu <b>ciągłego</b> - zbiór możliwych realizacji jest mocy <i>continuum</i> (nieskończony i nieprzeliczalny) (wzrost, waga)

Dla zmiennej losowej typu skokowego definiujemy funkcję rozkładu prawdopodobieństwa (krótko - rozkład prawdopodobieństwa), natomiast dla zmiennej losowej typu ciągłego wprowadzamy pojęcie funkcji gęstości prawdopodobieństwa (krótko - gęstość prawdopodobieństwa).

<b>Funkcja rozkładu prawdopodobieństwa (rozkład prawdopodobieństwa)</b> - funkcja przyporządkowująca realizacjom zmiennej losowej $X$ odpowiadające im prawdopodobieństwa. Rozkład prawdopodobieństwa można definiować za pomocą wzoru, tabeli lub wykresu. 

<b>Funckja gęstości prawdopodobieństwa (gęstość prawdopodobieństwa)</b> - dla zmiennej losowej typu ciągłego niemożliwe jest przypisanie wszystkim jej wartościom dodatnich prawdopodobieństw sumujących się do jedności. Można natomiast przypisać takie prawdopodobieństwa przedziałom liczbowym, np. przedziałowi $P(x<X<x+\Delta x)$, gdzie $\Delta x$ jest długością pewnego krótkiego przedziału o początku w punkcie $x$. Formalnie funkcja gęstpści prawdopodobieństwa jest zdefiniowana jako granica przy $\Delta x \rightarrow 0$, czyli

\begin{equation}
f(x)=\lim\limits_{\Delta x \rightarrow 0} \frac{P(x < X < x+ \Delta x)}{\Delta x}
\end{equation}

Można powiedzieć, że gęstość prawdopodobieństwa zmiennej losowej $X$ to średnia "ilość prawdopodobieństwa" przypadająca na jednostkę długości przedziału $(x, x+ \Delta x)$, gdy długość tego przedziału dąży do zera. 
 
<b>Dystrybuanta zmiennej losowej $X$</b> - jest to funkcja $F(x)$ zmiennej rzeczywistej $x$ określonej wzorem:

$F(x) = P(X<x)$ dla każdego $x \in \mathbb{R}$ 

Znając rozkład (lub gęstość), zawsze możemy znaleźć dystrybuantę, a znając dystrybuantę, zawsze możemy znaleźć rozkład (lub gęstość).

<b>Własności dystrybuanty:</b>
- (a) przyjmuje wartości z przedziału $[0, 1]$, tzn. $0 \leq F(x) \leq 1$ dla $x \in (-\infty, \infty)$,
- (b) jest funkcją niemalejącą, tzn. dla $x_1 < x_2$ zawsze $F(x_1) \leq F(x_2)$,
- (c) jest funkcją lewostronnie ciągłą,
- (d) $F(-\infty)=0$, $F(+\infty) = 1$,
- (e) $P(a \leq X \leq b) = F(b) - F(a)$.

Własności b, c, d są konieczne i wystarczające do tego, że by funkcja $F(x)$ była dystrtybuantą. Znając dystrybuantę posiadamy kompletną wiedzę na temat zmiennej losowej. Często jest to wiedza zbyt obszerna, ponieważ interesują nas wybrane ważniejsze cechy rozkładu, które można opisać za pomocą kilku liczb. 

<b>Parametr rozkładu zmiennej losowej</b> - każda liczba, która w pewien sposób charakteryzuje zbiór wartości, jakie może przyjąć zmienna losowa. 

Największe znaczenie praktyczne mają dwie grupy parametrów:
- reprezentujące przeciętną (średnią) wielkość zmiennej losowej, do najważniejszych należy <b>wartość oczekiwana (wartość przeciętna, nadzieja matematyczna)</b>, która dostarcza informacji na temat średniego poziomu zmiennej losowej
- parametry informujące o rozrzucie wartości zmiennej losowej, do najważniejszych należą <b>wariancja</b> i <b>odchylenie standardowe</b>

# 2. Rozkłady zmiennej typu skokowego

<b>Funkcję rozkładu prawdopodobieństwa dla zmiennej losowej skokowej</b> można zapisać:

$P(X=x_i) = p_i$, gdzie $p_i \geq 0$, $\sum\limits_{i=1}^{\infty} p_i = 1$, $i=1, 2, 3, ...$

Gdzie $x_i$ - punkty skokowe zmiennej losowej $X$, $p_i$ - skoki zmiennej $X$.

<b>Dystrybuanta zmiennej losowej skokowej X</b> ma postać:

$F(x) = \sum\limits_{x_i<x} p_i$ dla $i=1, 2, ...$

<b>Wartość oczekiwana zmiennej losowej skokowej X</b> wyraża się wzorem:
    
$E(X) = \sum\limits_{i=1}^{\infty} x_i p_i$

Jeżeli powyższy szereg jest bezwględnie zbieżny, czyli $\sum\limits_{i=1}^{\infty} |x_i| p_i < +\infty$ to wartość oczekiwana charakteryzuje się następującymi własnościami:
- wartość oczekiwana stałej równa się tej stałej, czyli $E(C) = C$,
- wartość oczekiwana sumy dwóch zmiennych losowych $X$ i $Y$ jest równa sumie wartości oczekiwanych tych zmiennych, czyli $E(X+Y) = E(X) + E(Y)$. Tą własność można uogólnić na dowolną liczbę zmiennych losowych,
- wartość oczekiwana iloczynu dwóch niezależnych zmiennych losowych jest równa iloczynowi wartości oczekiwanych tych zmiennych, czyli $E(XY) = E(X) \cdot E(y)$, z czego wynika, że stałą można wyciągać przed znak wartości oczekiwanej, czyli $E(CX) = E(C) \cdot E(X) = CE(X)$ 

<b>Wariancja zmiennej losowej X zmiennej losowej typu skokowego</b> to liczba zdefiniowana w następujący sposób:

$D^2 (X) = \sum\limits_{i=1}^{\infty} [x_i - E(X)]^2 p_i = E[X-E(X)]^2$

Wariancja zmiennej losowej X jest to wartość oczekiwana kwadratu odchyleń wartości tej zmiennej od wartości oczekiwanej. Przy obliczaniu wariancji często posługujemy się wzorem: 

$D^2 (X) = E(X^2) - [E(X)]^2$

Można powiedzieć, że wariancja jest równa wartości oczekiwanej kwadratu zmiennej, pomniejszonej o kwadrat wartości oczekiwanej.

<b>Odchylenie standardowe zmiennej losowej typu skokowego</b> wyraża się wzorem:

$D(X) = \sqrt{D^2 (X)}$

<b>Własności wariancji (odchylenia standardowego)</b>:
- wariancja ze stałej jest równa zeru, czyli $D^2(C) = 0$ 
- wariancja iloczynu stałej i zmiennej losowej jest równa iloczynowi kwadratu stałej i wariancji tej zmiennej, czyli $D^2(CX) = C^2 D^2(X)$
- wariacja sumy(różnicy) dwóch niezależnych zmiennych losowych równa się sumie wariancji tych zmiennych, czyli $D^2 (X \pm Y) = D^2(X) + D^2(Y)$, gdy $X$ i $Y$ są niezależne.

<b>Rozkład dwumianowy</b> - dotyczy sytuacji, w której wynikiem doświadczenia może być jedno z dwóch wzajemnie wykluczających się zdarzeń, przy czym prawdopodobieństwa tych zdarzeń są w każdym z $n$ doświadczeń jednakowe. Wówczas prawdopodobieństwo sukcesu możemy oznaczyć przez $p$ natomiast porażki przez $q = 1-p$. Przykładem takiego zdarzenia może być rzut monetą. 

<b>Rozkład Poissona</b> - inaczej rozkład zdarzeń rzadkich. Rozkład Poissona ma postać $P(X=k) = \frac{\lambda^k}{k!} e^{-\lambda}, ~ k=0, 1, 2, ...$. Z kolei jego dystrybuanta wyraża się wzorem $F(x) = \sum\limits_{k \leq x} \frac{\lambda^k}{k!} e^{-\lambda}, ~ k=0, 1, 2, ...$. Jest szczególnym przypadkiem rozkładu dwumianowego zachodzącym wtedy, gdy prawdopodobieństwo sukcesu jest małe, a liczba realizacji na tyle duża, że $np=\lambda$ jest wielkością stałą. Wartość oczekiwana i wariancja zmiennej losowej $X$ wynoszą $E(X) = D^2(X) = \lambda$. Jest rozkładem prawostronnie skośnym, wraz ze wzrostem parametru $\lambda$ zbliża się do rozkładu symetrycznego. Jest rozkładem jednoparametrycznym - zależy tylko od wartości $\lambda$.
